### Results Discussion

The analysis revealed that most crimes exhibited seasonal fluctuations. These findings are consistent with a study by Shiode et al. (2023) on the seasonal characteristics of different types of crime in London, which also found that many crimes display periodic, seasonal variations in their volume, a phenomenon referred to as "crime seasonality" (Shiode et al., 2023). Therefore, when comparing these results with existing literature, it is evident that the observed patterns align with expected outcomes. The results are also consistent with the data exploration findings which showed that most crimes peaked during the summer months and gradually decreased as summer ended. 

This pattern can help the West Yorkshire Metropolitan Police determine the optimal times to increase vigilance and prepare for specific types of crimes. For example, the analysis indicated that anti-social behavior, criminal damage and arson, violence, and public order offenses are more likely to occur from April to August. Therefore, increasing police presence during this period could deter potential offenders.

However, this pattern is not consistent across all types of crime. While some crimes showed a steady increase, others increased gradually, and some remained constant during the period, indicating an irregular pattern for different crime types.

To further substantiate these findings, I employed a time series decomposition model to understand the patterns for forecasting and planning. I used boxplots to identify and remove outliers that could affect the model, and standardized the data to improve its quality, accuracy, and reliability. Using an additive model, I built the time series model based on the monthly counts of each crime type. The results were visualized using a line chart, as shown below.

```python
# Standardizing data

crime_count['Count'] = (crime_count['Count'] - crime_count['Count'].mean()) / crime_count['Count'].std() 
crime_count
```

```python
# Outlier Detection
sns.boxplot(crime_count["Count"])
plt.show()
```

```python
# outlier treatment for count
Q1 = crime_count.Count.quantile(0.25)
Q3 = crime_count.Count.quantile(0.75)
IQR = Q3 - Q1
crime_count = crime_count[(crime_count.Count >= Q1 - 1.5*IQR) & (crime_count.Count <= Q3 + 1.5*IQR)]
```

```python
# Re-checking for outliers
sns.boxplot(crime_count["Count"])
plt.show()
```

```python
# Building model

from statsmodels.tsa.seasonal import seasonal_decompose

Trend = seasonal_decompose(crime_count['Count'], model='Additive', period=1)
Trend.plot()
plt.show()
```
